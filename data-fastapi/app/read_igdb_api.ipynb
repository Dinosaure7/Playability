# Create the function

```python
def get_json(game_name):
    
    wrapper = IGDBWrapper("q3o5f9in83s5h07mkv23dn0ps5tvry", "fq9ucxht9dl2fz9vsnfx9sr7vtl98r")
    
    game_name = '"' + game_name + '"'
    
    # Request the API of the game
    result = wrapper.api_request(
                "games",
                f"fields name, genres, cover, rating, summary, storyline, platforms, involved_companies, videos; where name = {game_name};"
            )

    in_json = result.decode('utf8').replace("'", '"')
    data = json.loads(in_json)
    # if the game is not found, I set the variables as empty
    if len(data) == 0:
        game_id = ""
        summary = ""
        rating = ""
        genres = ""
        platforms = ""
        companies = ""
        cover = ""
        video = ""
        
        return game_id, summary, rating, genres, platforms, companies, cover, video
    
    # Otherwise I run the code
    else:
        # I gather the available information
        try:
            game_id = data[0]["id"]
        except:
            game_id = ""
        try:
            summary = data[0]["summary"]
        except:
            summary = ""
        try:
            rating = data[0]["rating"]
        except:
            rating = ""
        
        # Create the dictionary with all the genres
        try:
            gen = wrapper.api_request(
                        'genres',
                        'fields name; limit 100;'
                    )

            gen_json = gen.decode('utf8').replace("'", '"')
            data_gen = json.loads(gen_json)

            dict_genre = {}
            for ind in range(len(data_gen)):
                dict_genre[data_gen[ind]["id"]] = data_gen[ind]["name"]

            # Get the genre in list
            genres = []
            for genre in data[0]["genres"]:
                genres.append(dict_genre[genre])
        except:
            genres = ""
            
        # Create the dictionary with all the platforms
        try:
            plat = wrapper.api_request(
                        'platforms',
                        'fields name; limit 500;'
                    )

            plat_json = plat.decode('utf8').replace("'", '"')
            data_plat = json.loads(plat_json)

            dict_plat = {}
            for ind in range(len(data_plat)):
                dict_plat[data_plat[ind]["id"]] = data_plat[ind]["name"]
        
            # Get the platforms in list
            platforms = []
            for platform in data[0]["platforms"]:
                platforms.append(dict_plat[platform])
        except:
            platforms = ""
            
        # Get the companies
        try:
            companies = []
            for comp in data[0]["involved_companies"]:
                inv_comp = wrapper.api_request(
                            'involved_companies',
                            f'fields company; where id = {comp};'
                        )

                inv_comp_json = inv_comp.decode('utf8').replace("'", '"')
                data_inv_comp = json.loads(inv_comp_json)

                data_inv_comp[0]["company"]

                company = wrapper.api_request(
                            'companies',
                            f'fields name; where id = {data_inv_comp[0]["company"]};'
                        )

                comp_json = company.decode('utf8').replace("'", '"')
                data_comp = json.loads(comp_json)

                companies.append(data_comp[0]["name"])
        except:
            companies = ""
            
        #Get the cover
        try:
            cover_api = wrapper.api_request(
                        'covers',
                        f'fields url; where game = {data[0]["id"]};'
                    )

            cov_json = cover_api.decode('utf8').replace("'", '"')
            data_cov = json.loads(cov_json)

            cover = data_cov[0]["url"].replace("t_thumb", "t_cover_big")
        except:
            cover = ""
            
        # Get the video
        try:
            videos_api = wrapper.api_request(
                        'game_videos',
                        f'fields *; where game = {data[0]["id"]};'
                    )

            vid_json = videos_api.decode('utf8').replace("'", '"')
            data_vid = json.loads(vid_json)

            youtube = "https://www.youtube.com/watch?v="
            video = youtube + data_vid[-1]['video_id']
        
        except:
            video = ""
    
        return game_id, summary, rating, genres, platforms, companies, cover, video
```
